```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Read data from 'results.csv'
df = pd.read_csv('results.csv')

# Plotting using seaborn
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.lineplot(data=df.set_index('Encoding'), markers=True, dashes=False)

# Customize the plot
plt.title('F1 Score for Different Encodings across Datasets', fontsize=16)
plt.xlabel('Encoding', fontsize=12)
plt.ylabel('F1 Score', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Dataset', loc='center left', bbox_to_anchor=(1, 0.5))

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# plot ptb results in a bar chart
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x='Encoding', y='pl', data=df)

# Customize the plot
plt.title('F1 Score for Different Encodings on pl_spmrl Dataset', fontsize=16)
plt.xlabel('Encoding', fontsize=12)
plt.ylabel('F1 Score', fontsize=12)
plt.xticks(rotation=45, ha='right')

# set the y-axis limits
plt.ylim(85, 100)

# Show the plot
plt.tight_layout()
plt.show()
```
